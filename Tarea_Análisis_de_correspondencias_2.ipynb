## Conjunto Mushroom

Este conjunto de datos incluye descripciones de muestras hipotéticas correspondientes a 23 especies de hongos laminosos de las familias Agaricus y Lepiota. Cada especie se identifica como definitivamente comestible, definitivamente venenosa o de comestibilidad desconocida y no recomendada.
  
**Atributos del Dataset "Mushroom"**

| Nombre del Atributo | Nombre en el Dataset |
| :--- | :--- |
| **Clase**  | `class` |
| Forma del Sombrero | `cap-shape` |
| Superficie del Sombrero | `cap-surface` |
| Color del Sombrero | `cap-color` |
| Magulladuras | `bruises` |
| Olor | `odor` |
| Adherencia de Laminillas | `gill-attachment` |
| Espaciado de Laminillas | `gill-spacing` |
| Tamaño de Laminillas | `gill-size` |
| Color de Laminillas | `gill-color` |
| Forma del Tallo | `stalk-shape` |
| Raíz del Tallo | `stalk-root` |
| Superficie del Tallo (Arriba) | `stalk-surface-above-ring` |
| Superficie del Tallo (Abajo) | `stalk-surface-below-ring` |
| Color del Tallo (Arriba) | `stalk-color-above-ring` |
| Color del Tallo (Abajo) | `stalk-color-below-ring` |
| Tipo de Velo | `veil-type` |
| Color del Velo | `veil-color` |
| Número de Anillos | `ring-number` |
| Tipo de Anillo | `ring-type` |
| Color de Esporas | `spore-print-color` |
| Población | `population` |
| Hábitat | `habitat` |

#### Carga de datos

```python
mushroom = fetch_openml(name='mushroom', as_frame=True)
df_m = mushroom.frame
for col in df_m.columns:
    df_m[col] = df_m[col].astype('category')
df_m.head()
```

```python
df_m.shape
```

```python
df_m.dtypes
```

#### Análisis de correspondencias

```python
mca_m = prince.MCA(n_components=2, random_state=42)
mca_m = mca_m.fit(df_m)
coords_rows = mca_m.row_coordinates(df_m)
coords_cols = mca_m.column_coordinates(df_m)
```

#### Gráfico de dispersión categorías

En la gráfica de a continuación, se representan cada una de las categorías que contienen las columnas del dataset. Permite identificar las asociaciones fuertes entre variables que representan las característica que suelen aparecer juntas, al igual que las asociaciones débiles, es decir, aquellas que casi nunca se ven en conjunto. Se puede identificar patrones claros. La gran nube de puntos localizada en la parte inferior, sugiere que hay un conjunto de características comunes en y que describen al hongo más típico del conjunto de datos, stalk-surface-above-ring_f (0.6371675,0.525369) y spore-prin-color_k (0.7224842, 0.4691313) por nombrar algunas. En la parte inferior izquierda, vemos un conjunto de puntos que representan características fuertemente asociadas entre ellas, pero completamente alejadas de las comentadas anteriormente. Lo más destacable son los puntos de la parte superior derecha, esos 5 puntos representan características que muy seguramente aparecen juntas en los hongos, pero es casi imposible que aparezcan con las características de los otros cúmulos.

```python
df_plot = coords_cols.reset_index()
df_plot.columns = ['Etiqueta', 'Dim 1', 'Dim 2']

fig = px.scatter(df_plot,
                 x='Dim 1',
                 y='Dim 2',
                 title='ACM - Mushroom Dataset (Interactivo)',
                 hover_data=['Etiqueta']
                )

fig.update_traces(marker=dict(color='red'))

fig.update_layout(width=800, height=800,
                  xaxis=dict(gridcolor='lightgray'),
                  yaxis=dict(gridcolor='lightgray'))

fig.show()
```

#### Contribución de las columnas a las dimensiones

```python
print("Contribuciones de las columnas a las dimensiones:")
print(mca_m.column_contributions_)
```

Se lista las características que más aportaron para el análisis, es decir, aquellas que son más representativas y nos podrían permitir predecir qué tipo de hongo es. Se destacan el tipo de tallo k, el tipo de anillo, el odor fétido y las clases para la dimensión 1. Para la dimensión 2, las laminillas adheridas, el tallo color naranda, el hábitat y la población. Con lo anterior, podemos concluir que el eje 1 está relacionado con la toxicidad del hongo, pues las características están relacionadas con este tipo de especie.

```python
# Se guarda las contribuciones en un dataframe
contrib_df = mca_m.column_contributions_.rename(columns={0: 'Dim 1', 1: 'Dim 2'})
# Ordenamiento
contrib_dim1_sorted = contrib_df.sort_values(by='Dim 1', ascending=False)
contrib_dim2_sorted = contrib_df.sort_values(by='Dim 2', ascending=False)

print("Contribuciones MÁS importantes para la Dim 1 (Eje X)")
print(contrib_dim1_sorted.head(10)) # Mostramos el top 10


print("\nContribuciones MÁS importantes para la Dim 2 (Eje Y)")
print(contrib_dim2_sorted.head(10)) # Mostramos el top 10
```

#### Biplot categoría-clase

En este gráfico tenemos los mismo datos que en el gráfico anterior, con la mejora de la aparición de dos indicadores sobre la clase, un tache de color rojo para los venenosos y uno verde para los comestibles. Con esto, obtenemos que el eje de la dimensión 1 es importante para definir la toxicidad del hongo al tener a los venenosos del lado izquierdo del origen y no venenosos del lado derecho. También se observa que hay un cúmulo de características cercanas a venenosos, mismas que son distintivas para definir su toxicidad, es decir, que están estrechamente relacionadas. Lo mismo ocurre con los hongos comestibles.

```python
# Obtener coordenadas de la columna 'class'
class_coords = coords_cols.filter(like='class', axis=0)

# Coordenadas de otras categorías
other_coords = coords_cols.drop(class_coords.index)

# Convertir 'other_coords' a DataFrame
df_others = other_coords.reset_index()
df_others.columns = ['Etiqueta', 'Dim 1', 'Dim 2']
df_others['Tipo'] = 'Otra categoría'
df_others['Color'] = 'blue'
df_others['Tamaño'] = 1

# Convertir 'class_coords' a DataFrame
df_class = class_coords.reset_index()
df_class.columns = ['Etiqueta', 'Dim 1', 'Dim 2']
df_class['Tipo'] = 'Class'
# Asignar color según la etiqueta
df_class['Color'] = ['green' if 'e' in idx else 'red' for idx in df_class['Etiqueta']]
df_class['Tamaño'] = 5

df_plot = pd.concat([df_others, df_class], ignore_index=True)

fig = px.scatter(df_plot,
                 x='Dim 1',
                 y='Dim 2',
                 # --- Aquí está la magia ---
                 title='ACM - Mushroom Dataset (Interactivo)',
                 hover_data=['Etiqueta'],
                 color='Color',
                 size='Tamaño',
                 symbol='Tipo',

                 color_discrete_map={
                     'blue': 'blue',
                     'green': 'green',
                     'red': 'red'
                 },
                 symbol_map={
                     'Class': 'x',
                     'Otra categoría': 'circle'
                 }
                )

fig.update_layout(
    width=800,
    height=800,
    legend_title_text='Categoría',
    xaxis=dict(gridcolor='lightgray'),
    yaxis=dict(gridcolor='lightgray')
)

fig.update_traces(showlegend=False, selector=dict(name='blue'))
fig.update_traces(showlegend=False, selector=dict(name='red'))
fig.update_traces(showlegend=False, selector=dict(name='green'))

fig.show()
```

#### Biplot categorías - hongos individuales

En esta gráfica tenemos un nuevo elemento, una nube de puntos azules que representan cada registro de hongos. Hay dos nubes de hongos, la del lado izquierdo que es densa y oscura, coincidiendo perfectamente con la ubicación del distintivo de clase venenosa, y sus característica, es decir, una gran parte del dataset son venenosos. Por otro lado, en la parte derecha hay nubes más dispersas, lo que indica que los hongos comestibles son más variados.

```python
# Filas (hongos)
coords_rows = mca_m.row_coordinates(df_m)

# Columnas (categorías)
coords_cols = mca_m.column_coordinates(df_m)

# Crear la figura base (Hongos)
fig = px.scatter(
    coords_rows,
    x=0,
    y=1,
    title='Biplot ACM - Mushroom Dataset',
    hover_name=coords_rows.index
)

fig.update_traces(
    marker=dict(size=10, opacity=0.2),
    name='Hongos',
    selector=dict(mode='markers')
)

fig.add_trace(
    go.Scatter(
        x=coords_cols[0],
        y=coords_cols[1],
        mode='markers',
        name='Categorías',
        marker=dict(color='red', size=10),
        hovertext=coords_cols.index,
        hoverinfo='x+y+text'
    )
)

fig.update_layout(
    width=900,
    height=900,
    xaxis_title="Dim 1",
    yaxis_title="Dim 2",

    xaxis_range=[-1.8, 2.0],
    yaxis_range=[-0.8, 5.8],

    template='plotly_white',
    xaxis_showgrid=True,
    yaxis_showgrid=True,

    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    ),

    hovermode="closest"
)

fig.show()
```

## Conjunto Car

Hay 406 observaciones sobre las siguientes 8 variables: MPG (millas por galón), número de cilindros, cilindrada (pulgadas cúbicas), caballos de fuerza, peso del vehículo (libras), tiempo de aceleración de 0 a 60 mph (segundos), año del modelo (módulo 100) y origen del vehículo (1. estadounidense, 2. europeo, 3. japonés). Estos datos aparecen en siete páginas. También se proporcionan las etiquetas de los vehículos (tipos) en el mismo orden que las 8 variables, en siete páginas separadas. Los valores faltantes se marcan con signos de interrogación.

### Atributos del Dataset "Car Evaluation"

| Nombre del Atributo | Nombre en el Dataset (Asumido) |
| :--- | :--- |
| Precio de Compra | `buying` |
| Costo de Mantenimiento | `maint` |
| Número de Puertas | `doors` |
| Capacidad (Personas) | `persons` |
| Tamaño del Maletero | `lug_boot` |
| Seguridad | `safety` |

La clase P es la clase mayoritaria del conjunto, es decir, No aceptado, por lo que P se tomará como Problemático y N como No Problemático

#### Carga de datos

```python
car = fetch_openml(name='car', as_frame=True)
df_c = car.frame
df_c.head()
```

```python
df_c['binaryClass'].unique()
```

```python
df_c.shape
```

```python
df_c.dtypes
```

#### Análisis de correspondencias

```python
mca_c = prince.MCA(n_components=2, random_state=42)
mca_c = mca_c.fit(df_c)
```

#### Contribución de las columnas a las dimensiones

```python
print("Contribuciones de las columnas a las dimensiones:")
print(mca_c.column_contributions_)
```

Se lista las características que más aportaron para el análisis, es decir, aquellas que son más representativas y nos podrían permitir predecirsi el auto tuvo o no aceptación. Se destacan la clasificación, la seguridad baja y la ocupación de dos personas para la dimensión 1. Para la dimensión 2, la ocupación 4 personas, más de 5 puertas, y el maletero pequeño. Con lo anterior, podemos concluir que el eje 1 está relacionado con la la aceptación del automovil, pues las características están relacionadas con ello.

```python
# Se guarda las contribuciones en un dataframe
contrib_df = mca_c.column_contributions_.rename(columns={0: 'Dim 1', 1: 'Dim 2'})
# Ordenamiento
contrib_dim1_sorted = contrib_df.sort_values(by='Dim 1', ascending=False)
contrib_dim2_sorted = contrib_df.sort_values(by='Dim 2', ascending=False)

print("Contribuciones MÁS importantes para la Dim 1 (Eje X)")
print(contrib_dim1_sorted.head(10)) # Mostramos el top 10


print("\nContribuciones MÁS importantes para la Dim 2 (Eje Y)")
print(contrib_dim2_sorted.head(10)) # Mostramos el top 10
```

#### Biplot características y clases

En esta gráfica se observan puntos de color azul que significan las diferentes categorías del dataset, y cruces rojas que hacen referencia a la clase. Es importante destacar que el eje de la Dimensión 1 pertenece a la aceptabilidad (relacionado con la variable objetivo), pues es claro que las dos X están en extremos opuestos del eje (izquierdo - aceptación positiva, derecho - aceptación negativa). En la zona derecha se encuentran las características relacionadas con lo autos con mala aceptación, como por ejemplo, seguridada baja o asientos para 2 personas, y en la zona izquierda con buena aceptación, como seguridad alta, 4 puertas. Parece ser un análisis obvio, intuitivamente estas características son las responsable de que un auto sea atractivo o no. Además, se observan grupos de puntos azules juntos, lo que representa que ciertas características suelen venir juntas, como maletero grande con 4 puertas.

```python
coords_cols = mca_c.column_coordinates(df_c)

df_plot = coords_cols.reset_index()
df_plot.columns = ['Etiqueta', 'Dim 1', 'Dim 2']

df_plot['Tipo'] = [
    'Clase' if (cat.startswith('binaryClass_') or cat.startswith('class_'))
    else 'Otra categoría'
    for cat in df_plot['Etiqueta']
]

fig = px.scatter(
    df_plot,
    x='Dim 1',
    y='Dim 2',
    title='Biplot ACM - Car Evaluation',
    hover_data=['Etiqueta'],

    color='Tipo',
    symbol='Tipo',

    color_discrete_map={
        'Clase': 'red',
        'Otra categoría': 'blue'
    },
    symbol_map={
        'Clase': 'x',
        'Otra categoría': 'circle'
    },

    size=[10 if tipo == 'Clase' else 6 for tipo in df_plot['Tipo']],
    size_max=12
)

fig.update_layout(
    width=900,
    height=700,
    template='plotly_white',
    xaxis_showgrid=True,
    yaxis_showgrid=True,
    legend_title_text='Categoría'
)

fig.show()
```

#### Biplot de cada automovil y características

En este biplot se añade un elemento, los automóviles. Se puede observar dos grandes nubes divididas claramente, siguiendo nuevamente la lógica del eje de la dimensión 1. Hay un grupo en la zona izquierda, mostrando que del dataset una gran cantidad de datos pertenecen a los coches aceptados y sus características son específicas. Del lado derecho encontramos otra nube, mucho más dispersa, lo que demuestra que los autos no aceptados son más variados, es decir, que muchas características son las que pueden afectar.

```python
coords_rows = mca_c.row_coordinates(df_c)
coords_cols = mca_c.column_coordinates(df_c)

df_plot_cols = coords_cols.reset_index()
df_plot_cols.columns = ['Etiqueta', 'Dim 1', 'Dim 2']

df_plot_cols['Tipo'] = ['Clase' if (cat.startswith('binaryClass_'))
                        else 'Otra categoría'
                        for cat in df_plot_cols['Etiqueta']]

df_plot_cols['Tamaño'] = [10 if tipo == 'Clase' else 6 for tipo in df_plot_cols['Tipo']]

fig = px.scatter(
    df_plot_cols,
    x='Dim 1',
    y='Dim 2',
    title='Biplot ACM - Car Evaluation (Completo: Carros y Categorías)',
    hover_data=['Etiqueta'],
    color='Tipo',
    symbol='Tipo',
    size='Tamaño',
    size_max=12,
    color_discrete_map={
        'Clase': 'red',
        'Otra categoría': 'blue'
    },
    symbol_map={
        'Clase': 'x',
        'Otra categoría': 'circle'
    }
)

fig.add_trace(
    go.Scatter(
        x=coords_rows[0],
        y=coords_rows[1],
        mode='markers',
        name='Carros (Individuos)',
        marker=dict(
            color='gray',
            size=5,
            opacity=0.2
        ),
        hoverinfo='none' )
)

fig.update_layout(
    width=900,
    height=700,
    template='plotly_white',
    xaxis_showgrid=True,
    yaxis_showgrid=True,
    legend_title_text='Leyenda'
)

fig.show()
```

#### Contribución de las columnas a las dimensiones

```python
print("Contribuciones de las columnas a las dimensiones:")
print(mca_a.column_contributions_)
```

Se lista las características que más aportaron para el análisis, es decir, aquellas que son más representativas y nos podrían permitir predecir qué tipo de ingreso se tiene. Se destacan el estado amoroso esposo, estado civil casado, la clase y el estado civil nunca casado para la dimensión 1. Para la dimensión 2, el género femenino, estado en el matrimonio divorcia, estado amoroso no casado, y la edad. Con lo anterior, podemos concluir que el eje 1 está relacionado con la el estado del salario, pues las características están relacionadas con ello.

```python
# Se guarda las contribuciones en un dataframe
contrib_df = mca_a.column_contributions_.rename(columns={0: 'Dim 1', 1: 'Dim 2'})
# Ordenamiento
contrib_dim1_sorted = contrib_df.sort_values(by='Dim 1', ascending=False)
contrib_dim2_sorted = contrib_df.sort_values(by='Dim 2', ascending=False)

print("Contribuciones MÁS importantes para la Dim 1 (Eje X)")
print(contrib_dim1_sorted.head(10)) # Mostramos el top 10


print("\nContribuciones MÁS importantes para la Dim 2 (Eje Y)")
print(contrib_dim2_sorted.head(10)) # Mostramos el top 10
```

### Biplot con características y clases

En esta gráfica, los puntos de color azul muestran todas las categorías del dataset, los X de color rojo las clases objetivos. El eje de la dimensión 1 es el relacionado con el ingreso, pues los puntos de clase están en los opuestos de él. En el lado izquierdo es posible encontrar el lado del <= a 50k, en este lado tenemos la mayor cantidad de categorías, mostrando que es el estado más común en el conjunto de datos con una gran variedad de características, por ejemplo, el estado civil nunca casado u ocupación otros servicios. En el lado derecho, tenemos los de mayores ingresos, con una serie de características más reducidas, es decir, más específicas. Al igual que en los otros conjuntos de datos, esto parece tener sentido, pues los salarios más atractivos son para perfiles con características determinadas.

```python
df_plot = coords_cols.reset_index()
df_plot.columns = ['Etiqueta', 'Dim 1', 'Dim 2']

df_plot['Tipo'] = [
    'Clase' if (cat.startswith('class_') or cat.startswith('income_'))
    else 'Otra categoría'
    for cat in df_plot['Etiqueta']
]

fig = px.scatter(
    df_plot,
    x='Dim 1',
    y='Dim 2',
    title='Biplot ACM - Adult Income Dataset',
    hover_data=['Etiqueta'],
    color='Tipo',
    symbol='Tipo',
    color_discrete_map={
        'Clase': 'red',
        'Otra categoría': 'blue'
    },
    symbol_map={
        'Clase': 'x',
        'Otra categoría': 'circle'
    },
    size=[10 if tipo == 'Clase' else 6 for tipo in df_plot['Tipo']],
    size_max=12
)

fig.update_layout(
    width=900,
    height=700,
    template='plotly_white',
    xaxis_showgrid=True,
    yaxis_showgrid=True,
    legend_title_text='Categoría'
)

fig.show()
```

#### Biplot para características e individuos

Con base en esta gráfica, es posible observar que la gran mayoría de individuos (puntos grises) pertenecen al perfil con ingresos inferiores a 50k, aunque también hay una nube de menor tamaño localizada en el lado derecho cerca al perfil de con ingresos superiores. Nuevamente, se puede concluir que el conjunto de ingresos inferiores tiene un gran número de combinaciones de características, mientras que el grupo de ingresos superiores es mucho más específicos y menos común.

```python
coords_rows = mca_a.row_coordinates(df_cat)
coords_cols = mca_a.column_coordinates(df_cat)

df_plot_cols = coords_cols.reset_index()
df_plot_cols.columns = ['Etiqueta', 'Dim 1', 'Dim 2']

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=coords_rows[0],
        y=coords_rows[1],
        mode='markers',
        name='Individuos',
        marker=dict(
            color='gray',
            size=5,
            opacity=0.1
        ),
        hoverinfo='skip'
    )
)

fig.add_trace(
    go.Scatter(
        x=df_plot_cols['Dim 1'],
        y=df_plot_cols['Dim 2'],
        mode='markers',
        name='Características',
        marker=dict(
            color='blue',
            size=8
        ),
        text=df_plot_cols['Etiqueta'],
        hovertemplate='<b>%{text}</b><br>Dim 1: %{x:.2f}<br>Dim 2: %{y:.2f}<extra></extra>'
    )
)

fig.update_layout(
    title='Biplot ACM - Adult Income',
    width=900,
    height=700,
    template='plotly_white',
    xaxis_title="Dim 1",
    yaxis_title="Dim 2",
    xaxis_showgrid=True,
    yaxis_showgrid=True,
    legend_title_text='Leyenda',
    hovermode='closest'
)

fig.show()
```

Conclusión: Esta técnica facilita el análisis de correlación de características, permitiendo que personas que no tienen conocimientos del área puedan comprenderlo.
